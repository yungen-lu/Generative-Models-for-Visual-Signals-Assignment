```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
baseline_psnr = np.load('results/baseline_psnr.npy')
baseline_psnr_gt = np.load('results/baseline_psnr_gt.npy')

baseline_ssim = np.load('results/baseline_ssim.npy')
baseline_ssim_gt = np.load('results/baseline_ssim_gt.npy')

uniform_psnr = np.load('results/uniform_psnr.npy')
uniform_psnr_gt = np.load('results/uniform_psnr_gt.npy')

uniform_ssim = np.load('results/uniform_ssim.npy')
uniform_ssim_gt = np.load('results/uniform_ssim_gt.npy')

sharp_linear_increase_psnr = np.load('results/sharp linear increase_psnr.npy')
sharp_linear_increase_psnr_gt = np.load('results/sharp linear increase_psnr_gt.npy')

sharp_linear_increase_ssim = np.load('results/sharp linear increase_ssim.npy')
sharp_linear_increase_ssim_gt = np.load('results/sharp linear increase_ssim_gt.npy')

log10_linspace_psnr = np.load('results/log10 linspace_psnr.npy')
log10_linspace_psnr_gt = np.load('results/log10 linspace_psnr_gt.npy')

log10_linspace_ssim = np.load('results/log10 linspace_ssim.npy')
log10_linspace_ssim_gt = np.load('results/log10 linspace_ssim_gt.npy')
```

```python
# compare psnr_gt
plt.figure(figsize=(12, 6))
plt.plot(baseline_psnr_gt, label='baseline')
plt.plot(uniform_psnr_gt, label='uniform')
plt.plot(sharp_linear_increase_psnr_gt, label='sharp linear increase')
plt.plot(log10_linspace_psnr_gt, label='log10 linspace')
plt.legend()
plt.title('Compare PSNR(Ground Truth)')
plt.xlabel("Step (10 iter per step)")
plt.ylabel("PSNR")
```

```python
# compare SSIM_gt
plt.figure(figsize=(12, 6))
plt.plot(baseline_ssim_gt, label='baseline')
plt.plot(uniform_ssim_gt, label='uniform')
plt.plot(sharp_linear_increase_ssim_gt, label='sharp linear increase')
plt.plot(log10_linspace_ssim_gt, label='log10 linspace')
plt.legend()
plt.title('Compare SSIM(Ground Truth)')
plt.xlabel("Step (10 iter per step)")
plt.ylabel("SSIM_gt")
```
